```python
import os
import json
import shutil
from datetime import datetime
from ultralytics import YOLO

# All paths on Google Drive
DRIVE_DIR = '/content/drive/MyDrive/PotholeDetection'
DATASET_DIR = f'{DRIVE_DIR}/dataset'
RUNS_DIR = f'{DRIVE_DIR}/runs'
MODELS_DIR = f'{DRIVE_DIR}/models'
METRICS_DIR = f'{DRIVE_DIR}/metrics'

# Ensure directories exist
for d in [RUNS_DIR, MODELS_DIR, METRICS_DIR]:
    os.makedirs(d, exist_ok=True)

DATA_YAML = f'{DATASET_DIR}/data.yaml'
print(f"Dataset: {DATA_YAML}")
print(f"Exists: {os.path.exists(DATA_YAML)}")
print(f"Saving models to: {MODELS_DIR}")
print(f"Saving metrics to: {METRICS_DIR}")
```

---
## 2. Train YOLOv8s
**Est. Time:** 45-60 mins
Trains and saves immediately.

```python
MODEL_NAME = 'yolov8s'
WEIGHTS = 'yolov8s.pt'

print("="*50)
print(f"TRAINING {MODEL_NAME.upper()}")
print("="*50)

save_model_path = f'{MODELS_DIR}/{MODEL_NAME}_best.pt'
save_metrics_path = f'{METRICS_DIR}/{MODEL_NAME}_metrics.json'

# 1. TRAIN
model = YOLO(WEIGHTS)
model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,
    patience=15,
    cache=True,
    amp=True,
    workers=2,
    project=RUNS_DIR,
    name=MODEL_NAME,
    exist_ok=True
)

# 2. EVALUATE
print(f"\nðŸ“Š Evaluating {MODEL_NAME} on test set...")
metrics = model.val(data=DATA_YAML, split='test')

# 3. SAVE TO DRIVE IMMEDIATELY
local_weights = f'{RUNS_DIR}/{MODEL_NAME}/weights/best.pt'
if os.path.exists(local_weights):
    shutil.copy(local_weights, save_model_path)
    print(f"\nðŸ’¾  Model Saved to Drive: {save_model_path}")
else:
    print(f"\nâŒ ERROR: Could not find weights at {local_weights}")

metrics_data = {
    'model': MODEL_NAME,
    'timestamp': datetime.now().isoformat(),
    'mAP50': float(metrics.box.map50),
    'mAP50-95': float(metrics.box.map),
    'precision': float(metrics.box.mp),
    'recall': float(metrics.box.mr),
}

with open(save_metrics_path, 'w') as f:
    json.dump(metrics_data, f, indent=2)
print(f"ðŸ’¾  Metrics Saved to Drive: {save_metrics_path}")
print(json.dumps(metrics_data, indent=2))
```

## 7. Compare All Models

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all saved metrics
all_metrics = []
for name in ['yolov8s', 'yolov9c', 'yolov10s', 'yolov11s']:
    path = f'{METRICS_DIR}/{name}_metrics.json'
    if os.path.exists(path):
        with open(path) as f:
            all_metrics.append(json.load(f))
    else:
        print(f"âš ï¸ {name}: Not found")

if all_metrics:
    df = pd.DataFrame(all_metrics)[['model', 'mAP50', 'mAP50-95', 'precision', 'recall']]
    print("\n" + "="*60)
    print("MODEL COMPARISON")
    print("="*60)
    print(df.to_string(index=False))
    
    # Find best
    best = max(all_metrics, key=lambda x: x['mAP50'])
    print(f"\nðŸ† BEST: {best['model']} (mAP50: {best['mAP50']:.4f})")
```

```python
# Save comparison chart
if len(all_metrics) >= 2:
    fig, ax = plt.subplots(figsize=(10, 5))
    x = [m['model'] for m in all_metrics]
    y = [m['mAP50'] for m in all_metrics]
    bars = ax.bar(x, y, color=['#4CAF50', '#2196F3', '#FF9800', '#9C27B0'][:len(x)])
    ax.set_ylabel('mAP@50')
    ax.set_title('Model Comparison')
    ax.set_ylim(0, 1)
    for bar, val in zip(bars, y):
        ax.text(bar.get_x() + bar.get_width()/2, val + 0.02, f'{val:.3f}', ha='center')
    plt.tight_layout()
    plt.savefig(f'{DRIVE_DIR}/comparison_chart.png', dpi=150)
    plt.show()
    print(f"\nâœ… Chart saved: {DRIVE_DIR}/comparison_chart.png")
```

```python
# Copy best model to generic 'best_model.pt'
if all_metrics:
    best = max(all_metrics, key=lambda x: x['mAP50'])
    src = f"{MODELS_DIR}/{best['model']}_best.pt"
    dst = f"{MODELS_DIR}/best_model.pt"
    shutil.copy(src, dst)
    print(f"âœ… Best model copied to: {dst}")
    
    # Save final results
    with open(f'{DRIVE_DIR}/comparison_results.json', 'w') as f:
        json.dump({'best': best['model'], 'results': all_metrics}, f, indent=2)
    print(f"âœ… Global Results: {DRIVE_DIR}/comparison_results.json")
```
